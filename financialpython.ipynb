```python
# importing necessary libraries


import yfinance as yf
import pandas as pd
# from pandas_datareader import data as pdr

import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
```

```python
end_date = dt.datetime.now()

start_date = end_date - dt.timedelta(days = 365*5)
end_date, start_date
```

```python
stocks = ['MSFT', 'SPY', 'QQQ']
```

```python
# DOWNLOADING THE DATE FROM YAHOO FINANCE

df = yf.download(stocks, start = start_date, end = end_date)
```

```python
# first five values

df.head()
```

```python
adjusted_close_price = df['Close']
```

```python
adjusted_close_price.head()
```

```python
log_return = np.log(adjusted_close_price / adjusted_close_price.shift(1))
log_return.head()
```

```python
# calculate the cumulative return

cumulative_log_return = log_return.cumsum()
```

```python
cumulative_log_return.plot(title = "Cumulative Returns", figsize = (10,6))
```

```python
log_return.plot()
plt.show()
```
